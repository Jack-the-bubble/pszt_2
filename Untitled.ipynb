```python
# PSZT - Podstawy sztucznej inteligencji 
# Projekt 2 

# Importowanie bibliotek, tensorflow zawiera w sobie Keras

import tensorflow as tf  # Bilblioteka glebokiego uczenia.Tensors - multi-dimensional arrays
from sklearn.model_selection import StratifiedKFold # K-krotna walidacja krzyzowa
import numpy as num
import matplotlib.pyplot as plt
import pandas as pandas


sum_val_error=[] # Wektor przechowujacy bledy walidacji modelu w kazdej k walidacji
sum_val_acc=[] # Wektor przechowujacy precyzje walidacji modelu w kazdej k walidacji

sum_ucz_error=[] # Wektor przechowujacy bledy zbioru uczacego modelu
sum_ucz_acc=[] # Wektor przechowujacy bledy zbioru uczacego modelu

srednia_val_error=[]; # Srednia wyciagnieta ze wszystkich bledow walidacji modelu
srednia_val_acc=[]; # Srednia wyciagnieta ze wszystkich precyzji walidacji modelu

# Zmienne pomocnicze 
c=[];
cd=[];

# Przygotowywanie danych

# Zaladowanie bazy danych mnist zawierajacej 28x28 obrazkow recznie pisanych cyfr 0-9 wraz z etykietami
mnist = tf.keras.datasets.mnist  
# Dane sa rozpakowywane do zmiennych, trenujacych i testujacych
(X_train, Y_train),(x_test, y_test) = mnist.load_data()  

# Wykonanie walidacji k - krotnej krzyzowej dla k = n_splits
# Oryginalna próba jest dzielona na K podzbiorów. 
# Każdy z nich bierze się jako zbiór testowy, a pozostałe razem jako zbiór uczący i wykonuje analizę. 
# Analiza jest więc wykonywana K razy. 
# K rezultatów jest następnie uśrednianych. 
skf = StratifiedKFold(n_splits=5)

# Oryginalna próba jest dzielona na K podzbiorów. 
# Każdy z nich bierze się jako zbiór testowy, a pozostałe razem jako zbiór uczący i wykonuje analizę. 
# Analiza jest więc wykonywana K razy. 
# K rezultatów jest następnie uśrednianych. 
for train_index, test_index in skf.split(X_train, Y_train):
    x_train, x_test = X_train[train_index], X_train[test_index]
    y_train, y_test = Y_train[train_index], Y_train[test_index]


    #plt.imshow(x_train[222],cmap=plt.cm.binary)
    #plt.show()
    #print(y_train[222])

    # Normalizacja danych 
    x_train = tf.keras.utils.normalize(x_train, axis=1)  
    x_test = tf.keras.utils.normalize(x_test, axis=1)  

    
    # Budowanie modelu sekwencyjnego, z wyprzedzeniem (feed-forward)
    model = tf.keras.models.Sequential()
    # Przetworzenie warstwy wielowymiarowej 28x28 na plaska 1x784
    model.add(tf.keras.layers.Flatten())  
    # Dodanie dwóch warstw ukrytych aktywowanych funkcja ReLu, 128 jednostek
    model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))  
    model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))  
    # Dodanie warstwy wyjsciowej 10 jednostek 10 klas, funkcja Softmax - rozkładu prawdopodobieństwa 
    model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))  

    #Kompilacja modelu 
    model.compile(optimizer='adam',  # Wybor optymalizatora
              loss='sparse_categorical_crossentropy',  # Sposob obliczania miary strat
              metrics=['accuracy'])  # Co jest sledzone 
    
    # Trenowanie modelu przez 3 epoki
    
    History=model.fit(x_train,y_train,epochs=3)  

    
   # Wyciagniecie danych historycznych ze struktury  
   
    cd=History.history
    c =History.history['loss']
    d=History.history['acc']
    
    sum_ucz_error.append(c[2])
    sum_ucz_acc.append(d[2])
    
    
    print("Cala historia")
    print(cd)
    

    # Przetestowanie modelu, ocena danych próbki z modelem
    
    val_loss, val_acc = model.evaluate(x_test, y_test) 
    
    
    # Błąd walidacji modelu
    #print(val_loss)  
    # Precyzja walidacji modelu
    #print(val_acc)  
    
    #Sumowanie błędów walidacji w wektor
    sum_val_error.append(val_loss)
    
    #Sumowanie precyzji walidacji w wektor
    sum_val_acc.append(val_acc)
    
    
# Wyswietlenie wektorow
    
print("Wektor bledow walidacji")
print(sum_val_error)
print("Wektor precyzji walidacji")
print(sum_val_acc)

print("Wektor bledow uczacych")
print(sum_ucz_error)
print("Wektor precyzji uczacych")
print(sum_ucz_acc)


# Rysowanie wykresow 

plt.plot(sum_val_error)
plt.plot(sum_ucz_error)
plt.show()

plt.plot(sum_val_acc)
plt.plot(sum_ucz_acc)
plt.show()

# Wyliczenie wyniku koncowego

srednia_val_error=num.mean(sum_val_error)
#print(srednia_val_error)
srednia_val_acc=num.mean(sum_val_acc)
#print(srednia_val_acc)


raw_data ={'wal_err':sum_val_error,'ucz_err':sum_ucz_error,'wal_acc':sum_val_acc,'ucz_acc':sum_ucz_acc,'srednia_val_acc':srednia_val_acc,'srednia_val_error':srednia_val_error}

        #  return x,y
df=pandas.DataFrame(raw_data, columns=['wal_err','ucz_err','wal_acc','ucz_acc','srednia_val_acc','srednia_val_error'])
df.to_excel("/Users/janzgorzelskiJohnny/Desktop/dane.xlsx",sheet_name='Arkusz1')
```
